# 2. Clustering Algorithm Implementation

## A) KMeans Clustering 
* KMeans Clustering is an unsupervised algorithm that groups similar data points into clusters based on their features.
* The Algorithm iteratively updates the centroids of the clusters and reassigns the data points to the closest cluster.
  
### Why KMeans Clustering might be suitable for teh Iris dataset?
* KMeans Clustering is suitable for the Iris dataset because it is a simple and efficient algorithm that can handle medium-sized datasets.
* And the Iris dataset has a relatively small number of features, which makes it easier to visualize and interpret the results.

### Apply KMeans clustering to the preprocessed Iris dataset and visualize the clusters.

```python
kmeans =KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
```

```python
# Visualize the clusters 
plt.figure(figsize=(8,6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=kmeans.labels_)
plt.title("KMeans Clustering")
plt.show()
```

## B) Hierarchical Clustering
* Hierarchical Clustering is an unsupervised learning algorithm that builds a hierarchy of clusters by merging or splitting existing Clusters.
* The algorithm can be visualized as a dendrogram, which show the hierarchy of clusters.

### Why Hierarchical Clustering might be suitable for the Iris dataset
* Hierarchical Clustering is suitable for the Iris dataset because it can handle datasets with a complex structure.
* The Dendrogram visualization can provide insights into the relationships between the clusters.


### Apply Hierarchical clustering to the preprocessed Iris dataset and visualize the clusters.

```python
hier_clust = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
hier_clust.fit(df)
```

```python
# Visualize the clsuters 
plt.figure(figsize=(8,6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=hier_clust.labels_)
plt.title("Hierarchical Clustering")
plt.show()
```
